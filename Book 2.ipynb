```python
import numpy as np                                       #Importing numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Data I-A.csv")
duration = ['2013-14','2014-15','2015-16']
gsdp = ['GSDP - CURRENT PRICES (` in Crore)']
percent_growth = ['(% Growth over previous year)']
df = df.loc[~df['Duration'].isin(['2016-17'])]
```

```python
df_avg_percent_growth = df.loc[df['Duration'].isin(duration)]
df_avg_percent_growth = df_avg_percent_growth.loc[df_avg_percent_growth['Items  Description'].isin(percent_growth)]
df_avg_percent_growth.fillna(df_avg_percent_growth.mean(),inplace=True)
df_avg_percent_growth.loc['Average Growth of States'] = df_avg_percent_growth.mean()
df_avg_percent_growth = df_avg_percent_growth.loc[['Average Growth of States']]
df_avg_percent_growth.dropna(axis='columns',inplace=True)
df_avg_percent_growth.drop(['All_India GDP'], axis=1,inplace=True)
df_avg_percent_growth.sort_values(by=['Average Growth of States'], axis=1,ascending=False,inplace=True)
```

```python
df_avg_percent_growth.loc['Average Growth of States'].plot(kind="bar",title ="Comparison of Average Growth of States over the year 2013-14, 2014-15 and 2015-16",figsize=(10,5))
plt.xlabel('States')
plt.ylabel('Average Growth of States')
```

```python
df_growth_overyear = df.loc[df['Items  Description'].isin(percent_growth)]
df_growth_overyear = df_growth_overyear.drop(['Items  Description'], axis=1)
df_growth_overyear.fillna(df_growth_overyear.mean(),inplace=True)
df_growth_overyear.dropna(axis='columns',inplace=True)
x = df_growth_overyear['Duration']
```
